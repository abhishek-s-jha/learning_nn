```python
import os
import cv2
import pandas as pd
import shutil
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
pd.set_option("display.max_rows", None)
```

```python
dataset = pd.read_csv("./dataset.csv", header=None)
```

```python
dataset
```

```python
# first run which resulted in 282 images

name = []
quality = []
height = []
width = []
channel = []

for value in tqdm(dataset.values):
#     if i == 1:
    img_loc = value[0]
    value = value[1]
    if img_loc.split("/")[0].split("_")[1][0] == 'h':
        img_name = img_loc.split("/")[1]
        img_type = "HDR"
        img_loc = os.path.join("./dataset/hdr/", img_loc)
        img = cv2.imread(img_loc)
    else:
        img_name = img_loc.split("/")[1]
        img_type = "NORMAL"
        img_loc = os.path.join("./dataset/normal/", img_loc)
        img = cv2.imread(img_loc)

    name.append(img_name)
    quality.append(img_type)
    height.append(img.shape[0])
    width.append(img.shape[1])
    channel.append(img.shape[2])

    title = "Name : "+img_name+"\n"+"Type : "+img_type+"\n"+"Shape : "+str(img.shape)
    plt.title(title)
    plt.imshow(img)
    plt.savefig(os.path.join("./visualization/",img_type+"_"+img_name),bbox_inches='tight',facecolor='white')
    plt.close()
#     print("image location : ",img_loc)
#     print("value : ",value)
#         break
```

```python
# second run which resulted in 604 images

name = []
quality = []
height = []
width = []
channel = []

for value in tqdm(dataset.values):
#     if i == 1:
    img_loc = value[0]
    value = value[1]
    if img_loc.split("/")[0].split("_")[1][0] == 'h':
        img_name = img_loc.split("/")[1]
        img_type = "HDR"
        img_loc = os.path.join("./dataset/hdr/", img_loc)
        img = cv2.imread(img_loc)
    else:
        img_name = img_loc.split("/")[1]
        img_type = "NORMAL"
        img_loc = os.path.join("./dataset/normal/", img_loc)
        img = cv2.imread(img_loc)

    name.append(img_name)
    quality.append(img_type)
    height.append(img.shape[0])
    width.append(img.shape[1])
    channel.append(img.shape[2])

    title = "Name : "+img_name+"\n"+"Type : "+img_type+"\n"+"Shape : "+str(img.shape)
    plt.title(title)
    plt.imshow(img)
    plt.savefig(os.path.join("./visualization/",img_type+"_"+value+img_name[-4:]),bbox_inches='tight',facecolor='white')
    plt.close()
#     print("image location : ",img_loc)
#     print("value : ",value)
#         break
```

```python
dataset['name'] = name
dataset['quality'] = quality
dataset['height'] = height
dataset['width'] = width
dataset['channel'] = channel
```
